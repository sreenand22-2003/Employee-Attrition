### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
def Zeroone(x):
    if x=="Yes":
        return 1
    elif x=="No":
        return 0
```

```python
df["Attrition"]=df["Attrition"].apply(Zeroone)
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf= pd.get_dummies(df,drop_first=True,dtype="int")
```

```python
newdf.head()
```

```python
newdf.shape
```

```python
for i in cate2:
    unique = df[i].unique()
    print("unique values in ",i,"are:",unique)
```

#### Split the columns into input and target variables (X and y)

```python
x=newdf.drop(columns=['Attrition'])
y=newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns=x.columns
```

```python
x=scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x,columns=xcolumns)
```

```python
x.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("accuracy score of our model is : ",metrics.accuracy_score(y_test, y_pred))
print("confusion matrics is : \n",metrics.confusion_matrix(y_test, y_pred))
print("classification report is : \n",metrics.classification_report(y_test, y_pred))
print("cros validation score is : ",cross_val_score(LogisticRegression(), x, y, cv = 3).max())
```

<span style = "color:orange"> Visualize confusion matrix </span>
